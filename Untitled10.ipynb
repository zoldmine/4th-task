```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('2016.csv')
Year_2016 = data['atccode1'].value_counts(ascending=False)[:10]
Year_2016
```

```python
fig, ax = plt.subplots()
Year_2016.plot(kind='barh')
ax.set(title='ТОП-10 Перечнь жизненно необходимых и важнейших лекарственных препаратов для медицинского применения за 2016 год',
xlabel = 'Количество, шт', ylabel='категория препарата')
```
